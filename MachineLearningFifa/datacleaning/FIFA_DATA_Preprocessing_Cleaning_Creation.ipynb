# Creating a Scaled Score Metric 

The scale_score metric is a metric created with the purpose of scoring young top prospects that are of bargain value. The idea behind it is to identify young prospects for clubs and buy them while they are relatively cheap. The scale score penalizes age and player value on the transfer market but maximizes overall and potential rating.
<br/>
The formula to create scale score is based on two new created metrics: adjGrade and divisor.
The calculation for these metrics is as follows:
<br/>
adjGrade:


<img src="https://latex.codecogs.com/gif.latex?adjGrade=&space;\left&space;[&space;(potential(\sqrt{(overall)(potential-overall)})))&space;\ast&space;\frac{(potential&plus;overall)}{2}&space;\right&space;]&space;&plus;&space;\sigma&space;_{eurvalue}" title="adjGrade= \left [ (potential(\sqrt{(overall)(potential-overall)}))) \ast \frac{(potential+overall)}{2} \right ] + \sigma _{eurvalue}" />

divisor:

<img src="https://latex.codecogs.com/gif.latex?divisor=&space;Age&space;\times&space;(eurvalue&space;&plus;&space;\sigma&space;_{eurvalue})" title="divisor= Age \times (eurvalue + \sigma _{eurvalue})" />

Then both adjGrade and divisor are run through standard scaling with the standard scaler from sklearn, and our results are stored as "adj_Grade_scale" and "div_scale". 
<br/> Thus, our scale score is the quotient of adj_Grade_scale/div_scale.

<img src="https://latex.codecogs.com/gif.latex?scale&space;score=\frac{adjGradeScale}{divScale}" title="scale score=\frac{adjGradeScale}{divScale}" />

```python
df= pd.read_csv("Full_DF.csv")

data=df.copy()
def pot_ovr_diff (row):
    return row['Potential'] - row['Overall'] 
data['pot_ovr_diff'] = data.apply(pot_ovr_diff, axis=1)
```

```python
data=data.copy()
def ovrXdiff (row):
    return row['Overall'] *row['pot_ovr_diff'] 
data['ovrXdiff'] = data.apply(ovrXdiff, axis=1)
```

```python
df2=pd.DataFrame(data["ovrXdiff"]).apply(np.sqrt)
df2.columns = ['ovrXDiffSqr']
complete = pd.concat([df2, df], axis=1)
data=complete.copy()

```

```python
data=complete.copy()
def potX (row):
    return row['Potential'] *row['ovrXDiffSqr'] 
data['potX'] = data.apply(potX, axis=1)
```

```python
data=data.copy()
def avgGrade (row):
    return ((row['Potential'] + row['Overall']))/2 
data['avgGrade'] = data.apply(avgGrade, axis=1)
```

```python
data.head()
```

```python
data=data.copy()
def grade2 (row):
    return row['potX'] *row['avgGrade'] 
data['grade2'] = data.apply(grade2, axis=1)
```

```python
data=data.copy()
def div (row):
    return row['age'] *(row['eur_value']+58693.359388)
data['div'] = data.apply(div, axis=1)
```

```python
std=data[['grade2']].std()
std.head()
data=data.copy()
def adjGrade (row):
    return row['grade2'] +(row['eur_value']+58693.359388)
data['adjGrade'] = data.apply(adjGrade, axis=1)
```

```python
data=data.copy()
def score (row):
    return row['adjGrade'] /row['div'] 
data['score'] = data.apply(score, axis=1)
```

```python
X=data[["div", "adjGrade", "score"]]
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std=sc.fit_transform(X)
X_std=pd.DataFrame(X_std)

X_std.columns = ['div_scale', 'adjGrade_scale', 'raw_score']

df3=X_std.copy()
def scale_score (row):
    return row['adjGrade_scale'] /row['div_scale'] 
df3['scale_score'] = df3.apply(scale_score, axis=1)
```

```python
complete = pd.concat([complete, df3], axis=1)
complete.head()
```

```python
complete.to_csv("scoring")
writer = pd.ExcelWriter('output.xlsx')
complete.to_excel(writer,'scoring')
writer.save()
```

# Creating a Value Grade

Here we convert all Potential and overall ratings to grades on a scale of 0-100 where the highest ratings are 100 and the lowest are graded at zero. We do the same for age and euro value except we do the converse. The highest age and euro_value have a grade of zero and the lowest have a score of 100. Then we generate a metric called grade that is the average of these scores.
<br>


down vote
Any variable (univariate distribution) v with observed minold and maxold values (or these could be preset potential bounds for values) can be rescaled to range minnew to maxnew by the following formula:

maxnew−minnewmaxold−minold⋅(v−maxold)+maxnew
or

maxnew−minnewmaxold−minold⋅(v−minold)+minnew.

```python
df= pd.read_csv("full_fifa18_data.csv")
df.describe().transpose()
```

```python
data=df.copy()
f1=(100/48)
def pot (row):
    return ((row['Potential']-94)* f1)+100
data['pot'] = data.apply(pot, axis=1)
```

```python
data=data.copy()
f2=(100/48)
def ovr (row):
    return ((row['Overall']-94)* f2)+100
data['ovr'] = data.apply(ovr, axis=1)
```

```python
data=data.copy()
f3=(100/31)
def age_grade (row):
    return 100-(((row['age']-47)* f3)+100)
data['age_grade'] = data.apply(age_grade, axis=1)
```

```python
data=data.copy()
f4=(100/123000000)
def val_grade (row):
    return 100-(((row['eur_value']-123000000)* f4)+100)
data['val_grade'] = data.apply(val_grade, axis=1)
```

```python
data=data.copy()
def grade (row):
    return (row['pot']+ row['ovr']+ row['age_grade'] +row['val_grade'])/4
data['grade'] = data.apply(grade, axis=1)
data.head()
```

```python
data.to_csv("grades")
writer = pd.ExcelWriter('output2.xlsx')
data.to_excel(writer,'grades')
writer.save()
```
